```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
print(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Build URL query for each city
url = "http://api.openweathermap.org/data/2.5/weather?"
for city in cities:
    # Build query URL
    query_url = f"{url}appid={weather_api_key}&q={city}"

#Test json dump for last city in query...
weather_response = requests.get(query_url)
weather_json = weather_response.json()
print(json.dumps(weather_json, indent=4))
```

```python
#Initiate sets
city_name = []
country = []
lat = []
lon = []
temp = []
humidity = []
cloudiness = []
wind = []
date = []

#Process through each city and collect data in lists through try/except

for (index, city) in enumerate(cities):
    try:
        query_url = f"{url}appid={weather_api_key}&q={city}"
        weather_response = requests.get(query_url)
        weather_json = weather_response.json()
        city_name.append(weather_json["name"])
        country.append(weather_json["sys"]["country"])
        lat.append(weather_json["coord"]["lat"])
        lon.append(weather_json["coord"]["lon"])
        temp.append(weather_json["main"]["temp"])
        humidity.append(weather_json["main"]["humidity"])
        cloudiness.append(weather_json["clouds"]["all"])
        wind.append(weather_json["wind"]["speed"])
        date.append(weather_json["dt"])


        print(f'Processing # {index} : {city}.')

    except:
        print("Not found.")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a data frame from cities + weather info
weather_dict = {
    "city": city_name,
    "country": country,
    "lat": lat,
    "lon":lon,
    "temp": temp,
    "humidity": humidity,
    "cloudiness": cloudiness,
    "windiness": wind,
    "datetime": date
}
weather_data = pd.DataFrame(data=weather_dict)
weather_data.head()
```

```python
# Export csv
weather_data.to_csv("./weather_data.csv")
```

## Linear Regression

```python
# create northern hem. and southern hem. dataframes
northern = weather_data
southern = weather_data
northern = northern.loc[weather_data["lat"] >= 0]
southern = southern.loc[weather_data["lat"] < 0]
southern.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern['lat']
y_values = northern['temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = f"y= {slope:.2f}x {'+' if intercept > 0 else ''} {intercept:.2f}"
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(50,300),fontsize=12,color="red")
plt.annotate(f'r ={str(rvalue)}',(50,290),fontsize=12,color="red")
plt.xlabel('Latitude')
plt.ylabel('Temp (C)')

# Save the figure
plt.savefig("north_temp_regression.png")

plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = southern['lat']
y_values = southern['temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = f"y= {slope:.2f}x {'+' if intercept > 0 else ''} {intercept:.2f}"
plt.title("Southern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(-20,280),fontsize=12,color="red")
plt.annotate(f'r ={str(rvalue)}',(-20,278),fontsize=12,color="red")
plt.xlabel('Latitude')
plt.ylabel('Temp (C)')

# Save the figure
plt.savefig("south_temp_regression.png")

plt.show()
```

Using linear regression, we are able to see a clear correlation between latitude and temperature. This is especially clear looking at the northern hemisphere data, where the r-value is 0.89. In both charts, temperature decreases away from the equator.
